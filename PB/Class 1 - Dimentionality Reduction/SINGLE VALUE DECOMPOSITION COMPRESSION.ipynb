# SINGLE VALUE DECOMPOSITION COMPRESSION

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
img = imread('Lenna.png')
X   = np.mean(img, -1) #CONVERTS TO GRAYSCALE

plt.imshow(img)
plt.axis(False)
```

Probability Distribution

```python
hist = list()
for linha in X:
    for i in linha:
        hist.append(i)
shush = plt.hist(hist, density=True, bins=100)
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False) #full_matrices=False saves computation time
#S        = np.diag(S) #MAKES FULL MATRIX
```

```python
plt.imshow(U @ np.diag(S) @ VT) # @ -> matrix multiplication
```

We can see that the lowest $\delta$ values are the ones with the biggest weight

```python
plt.plot(S)
plt.yscale('log')
```

```python
new_rank = 60

new_X    = U[:,:new_rank] @ np.diag(S)[:new_rank, :new_rank] @ VT[:new_rank, :]
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

plt.imshow(new_X, cmap='gray')
plt.axis(False)
```
